```python
tokens_amount = []
with open('generated_pushkin.txt', 'a', encoding='utf-8') as f:
    for i in tqdm(range(500)):
    
        completion = client.chat.completions.create(
          model="gpt-4o-mini",
          messages=[
            {"role": "system", "content": "Ты оссистент, который любит писать рассказы в стиле Александра Сергеевича Пушкина. Используешь только русский язык и пишешь интересно. Пишешь на неожиданные темы"},
            {"role": "user", "content": "Придумай интересный рассказ на случайную тему в пределах 3000 символов в стиле Александра Сергеевечиа. Мне нужен только текст рассказа, без названий, сносок."}
          ]
        )
        tokens_amount += [completion.usage.total_tokens]

        print(completion.choices[0].message.content + '#' * 10, file=f)
```

```python
print("tokens per text =", sum(tokens_amount) / len(tokens_amount))
```
