```python
import nltk
from pymystem3 import Mystem
from nltk.tokenize import word_tokenize
from collections import Counter
import matplotlib.pyplot as plt
```

### лемматизация текста

```python
nltk.download('punkt')
```

```python
mystem = Mystem()
```

```python
with open('../dataset/pushkin/cleaned.txt', encoding='utf-8') as f:
    text = f.read()
```

```python
text[:200]
```

```python
tokens = word_tokenize(text, language='russian')
```

```python
tokens
```

```python
lemmatized_text = mystem.lemmatize(' '.join(tokens))
```

```python
lemmatized_text
```

```python
lem_text = ''.join(lemmatized_text)
```

```python
lem_text[:100]
```

```python
orig_words = word_tokenize(text, language='russian')
lem_words = word_tokenize(lem_text, language='russian')
print('количетво разных слов в оригинальном тексте:', len(set(orig_words)))
print('количетво разных слов в лемматизированом тексте:', len(set(lem_words)))
```

```python
orig_cou = Counter(orig_words)
lem_cou = Counter(lem_words)

orig_rate = sorted([i[1] for i in orig_cou.items()])
lem_rate = sorted([i[1] for i in lem_cou.items()])
```

```python
plt.figure(figsize=(16, 9))
plt.plot(orig_rate, label='orig_words')
plt.plot(lem_rate, label='lemmed_words')
plt.xlabel('слово')
plt.ylabel('раз встретилось в тексте')
plt.title('частотное распределение оригинальных и лемматизированых слов')
plt.xlim((0, len(orig_rate) + 1000))
plt.ylim((-10, 1_000))
plt.legend()
plt.show()
```
