```python
sequences = []
sequence_lenght = 30
for i in range(sequence_lenght, len(words)):
    sequence = words[i-sequence_lenght:i]
    target_word = words[i]
    sequences.append((sequence, target_word))
```

### model craeting

```python
class TextGenerationModel(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, num_layers):
        super(TextGenerationModel, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, batch_first=True, num_layers=num_layers)
        self.fc = nn.Linear(hidden_dim, vocab_size)
        
    def forward(self, x):
        x = self.embedding(x)
        lstm_out, _ = self.lstm(x)
        out = self.fc(lstm_out[:, -1, :])
        return out
        
```

```python
device = 'cuda'
```

```python
model = TextGenerationModel(vocab_size=len(vocab), embedding_dim=100, hidden_dim=257, num_layers=2).to(device)
optimizer = optim.Adam(model.parameters(), )
loss_fn = nn.CrossEntropyLoss()
```

```python
class TextDataset(Dataset):
    def __init__(self, sequenes, word_to_index):
        self.sequences = sequences
        self.word_to_index = word_to_index

    def __len__(self):
        return len(self.sequences)

    def __getitem__(self, idx):
        sequence, target_word = self.sequences[idx]
        sequence = torch.tensor([self.word_to_index[word]   for word in sequence], dtype=torch.long).to(device)
        target = torch.tensor(self.word_to_index[target_word], dtype=torch.long).to(device)
        return sequence, target


dataset = TextDataset(sequences, word_to_index)
data_loader = DataLoader(dataset, batch_size=32, shuffle=True)
```

```python
num_epochs = 1
losses = []
i = 0
for epoch in range(num_epochs):
    for batch in tqdm(data_loader):
        sequence, target = batch

        # Forward pass
        outputs = model(sequence)
        loss = loss_fn(outputs, target)

        # Backward pass and optimization
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        if i % 1000 == 0:
            losses.append(loss)
        i += 1
```

```python
real_losses = [losses[i].item() for i in range(len(losses))]

plt.plot(real_losses)
plt.xlabel('every 1000 iters')
plt.ylabel('losses')
plt.show()
```

```python
def generate_text(seed_text, model, vocab, word_to_index, index_to_word, sequence_length, gen_length):
    model.eval()

    seed = torch.tensor([word_to_index[word] for word in seed_text.split()], dtype=torch.long).unsqueeze(0)
    generated_text = seed_text

    for _ in range(gen_length):
        with torch.no_grad():
            output = model(seed)

        _, next_word_index = torch.max(output, dim=1)
        next_word = index_to_word[next_word_index.item()]
        generated_text += " " + next_word
        seed = torch.cat((seed, next_word_index.unsqueeze(0)), dim=1)

    return generated_text
```

```python
model.to('cpu')
seed_text = "буря мглою небо кроет вихри снежные крутя"
gen_length = 100  # generate 100 words after the seed text

# Creating a reverse dictionary that maps indices to words
index_to_word = {index: word for word, index in word_to_index.items()}

# Generating the text
generated_text = generate_text(seed_text, model, vocab, word_to_index, index_to_word, sequence_length=30, gen_length=gen_length)

print(generated_text)
```
