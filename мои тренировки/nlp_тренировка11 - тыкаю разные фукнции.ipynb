```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

def visualize_tensor(tensor):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    
    # Определяем размерности тензора
    x_dim, y_dim, z_dim = tensor.shape
    
    # Создаем сетку координат
    x, y, z = np.indices((x_dim, y_dim, z_dim))

    # Определяем позиции кубов внутри параллелепипеда
    x = x.flatten()
    y = y.flatten()
    z = z.flatten()

    # Определяем размеры кубов (все одинаковые)
    dx = dy = dz = 0.5
    
    # Определяем значения тензора (выравниваем до 1D массива)
    values = tensor.flatten()
    
    # Нормализуем значения для цветовой карты
    norm = plt.Normalize(values.min(), values.max())
    colors = plt.cm.viridis(norm(values))
    
    # Визуализируем тензор в виде параллелепипеда с цветовой картой
    ax.bar3d(x - 0.5, y - 0.5, z - 0.5, dx, dy, dz, color=colors, shade=True)
    
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_zlabel('Z axis')
    
    plt.show()

# Пример использования
tensor = np.random.rand(4, 40, 4)  # Создаем случайный 3D тензор
visualize_tensor(tensor)
```
