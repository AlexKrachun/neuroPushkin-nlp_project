```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer, TextDataset, DataCollatorForLanguageModeling, Trainer, TrainingArguments
```

```python
!ls
```

```python


device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")


model_name = "gpt2"
model = GPT2LMHeadModel.from_pretrained(model_name).to(device)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

def load_dataset(file_path, tokenizer, block_size=128):
    dataset = TextDataset(
        tokenizer=tokenizer,
        file_path=file_path,
        block_size=block_size,
    )
    return dataset

file_path = "../dataset/pushkin - full.txt"
train_dataset = load_dataset(file_path, tokenizer)


data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False,
)

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã –æ–±—É—á–µ–Ω–∏—è
training_args = TrainingArguments(
    output_dir="./results",
    overwrite_output_dir=True,
    num_train_epochs=2,
    per_device_train_batch_size=4,
    save_steps=10_000,
    save_total_limit=2,
    fp16=True,  # –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ 16-–±–∏—Ç–Ω—ã—Ö —á–∏—Å–µ–ª –¥–ª—è —É—Å–∫–æ—Ä–µ–Ω–∏—è –æ–±—É—á–µ–Ω–∏—è –Ω–∞ GPU
)

# –°–æ–∑–¥–∞–Ω–∏–µ —Ç—Ä–µ–Ω–µ—Ä–∞ –∏ –∑–∞–ø—É—Å–∫ –æ–±—É—á–µ–Ω–∏—è
trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=train_dataset,
)

```

```python
trainer.train()

output_dir = "./pushkin_model"

# –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –¥–æ–æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏ –≤ —É–∫–∞–∑–∞–Ω–Ω—É—é –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏—é
trainer.save_model(output_dir)
tokenizer.save_pretrained(output_dir)

```

```python
model_path = './pushkin_model'
gpt = 'gpt2'

# tokenizer = GPT2Tokenizer.from_pretrained(gpt)
# model = GPT2LMHeadModel.from_pretrained(gpt, pad_token_id=tokenizer.eos_token_id).to(device)

tokenizer = GPT2Tokenizer.from_pretrained(model_path)
model = GPT2LMHeadModel.from_pretrained(model_path, pad_token_id=tokenizer.eos_token_id).to(device)

```

```python
def generate_text(prompt):
    input_ids = tokenizer.encode(prompt, return_tensors='pt').to(device)
    output = model.generate(
        input_ids,
        min_length=512,
        max_length=1024,
        num_beams=5,
        no_repeat_ngram_size=2,
        top_k=50, 
        top_p=0.95, 
        early_stopping=True,
    )
    return tokenizer.decode(output[0], skip_special_tokens=True)
```

```python
prompt = "—à–≤–∞–±—Ä–∏–Ω –ø—Ä–æ–≥—É–ª–∏–≤–∞—è—Å –ø–æ"
generated_texts = generate_text(prompt)
print(generated_texts)
```
