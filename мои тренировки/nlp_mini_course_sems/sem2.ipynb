```python
x_valid = torch.linspace(-20, 20, 300)
y_valid = torch.cos(x_valid.data)
plt.plot(x_valid.numpy(), y_valid.numpy(), 'o')
```

```python
x_valid.unsqueeze_(1)
y_valid.unsqueeze_(1)
```

```python
class CosNetwork(torch.nn.Module):
  def __init__(self, n_hidd_n):
    super(CosNetwork, self).__init__()
    self.layer1 = torch.nn.Linear(1, n_hidd_n)
    self.act1 = torch.nn.Sigmoid()
    self.layer2 = torch.nn.Linear(n_hidd_n, n_hidd_n)
    self.act2 = torch.nn.Sigmoid()
    self.layer3 = torch.nn.Linear(n_hidd_n, 1)


  def forward(self, x):
    x = self.layer1(x)
    x = self.act1(x)
    x = self.layer2(x)
    x = self.act2(x)
    x = self.layer3(x)
    return x


my_nn = CosNetwork(10)
```

```python
def predict(net, x, y):
  y_pred = net.forward(x)
  plt.plot(x.numpy(), y.numpy(), 'o', c='g')
  plt.plot(x.numpy(), y_pred.data.numpy(), 'o', c='r')

predict(my_nn, x_valid, y_valid)
```

```python
optimizer = torch.optim.Adam(my_nn.parameters(), lr=0.001)
```

```python
def loss_f(pred, target):
  sq = (pred - target) ** 2
  return sq.mean()
```

```python
for ep_id in range(20_000):

  optimizer.zero_grad()

  y_pred = my_nn.forward(x_train)
  loss_val = loss_f(y_pred, y_train)

  loss_val.backward()
  optimizer.step()


```

```python
predict(my_nn, x_valid, y_valid)

```
