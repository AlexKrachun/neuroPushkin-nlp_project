```python
def parse(row):
  id_, title, genre, desc = row.split(':::')
  return desc, title
```

```python
descs = [parse(row)[0] for row in rows]
```

```python
vect = CountVectorizer(min_df=0.001, max_df=0.8)
vect.fit(descs)
```

```python
vect.transform(descs)
```

```python
# суть слова = корень
stemmer = PorterStemmer()
word = 'ate'
print(f'{word}->', stemmer.stem(word))
```

```python
import string
```

```python
def tokenize(text):
  tokens = word_tokenize(text)
  tokens = [token for token in tokens if token not in string.punctuation]
  return tokens
```

```python
tokenize('Hello, world!')
```

```python
def tokenize_stem(text):
  tokens = word_tokenize(text)
  tokens = [token for token in tokens if token not in string.punctuation]
  tokens = [stemmer.stem(token) for token in tokens]
  return tokens
```

```python
print(tokenize('hello, powerful world!'))
print(tokenize_stem('hello, powerful world!'))
```
