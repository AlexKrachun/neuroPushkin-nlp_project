```python
def show(objective,
         x_lims=[-10.0, 10.0],
         y_lims=[-10.0, 10.0],
         x_ticks=100,
         y_ticks=100):
    x_step = (x_lims[1] - x_lims[0]) / x_ticks
    y_step = (y_lims[1] - y_lims[0]) / y_ticks
    X, Y = np.mgrid[x_lims[0]:x_lims[1]:x_step, y_lims[0]:y_lims[1]:y_step]
    res = []
    for x_id in range(X.shape[0]):
        res.append([])
        for y_id in range(X.shape[1]):
            x_val = X[x_id, y_id]
            y_val = Y[x_id, y_id]
            res[-1].append(objective(np.array([[x_val, y_val]]).T))
    res = np.array(res)
    plt.figure(figsize=(8, 8))
    plt.contour(X, Y, res, 100)
    plt.xlabel('$x_1$')
    plt.xlabel('$x_2$')
```

```python
def func_parab(var):
  return 10 * (var ** 2).sum()

def func_happy(var):
  return (var ** 1/2).sum()
```

```python
def grad_step(func, var):
  res = func(var)
  res.backward()
  var.data -= 0.01 * var.grad
  var.grad.zero_()

```

```python
x = torch.tensor(
    [7., 7.,], requires_grad=True
)

vars = []
fs = []
# optimizer = torch.optim.SGD([x], lr=0.001)

for i in range(100):
  vars.append(x.data.numpy().copy())
  fs.append(func_parab(x).data.cpu().numpy().copy())
  grad_step(func_parab, x)

```
