## выбираю размер словаря и делаю токенайзер

```python
from collections import Counter
import matplotlib.pyplot as plt

from tqdm.auto import tqdm

from pymystem3 import Mystem

from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.trainers import BpeTrainer
from tokenizers.pre_tokenizers import Whitespace

from transformers import GPT2Tokenizer


import re
```

## отчищу данные от разметочных символов для токенизации

```python
paths = ['generated_pushkin.txt', 'generated_curious_pushkin.txt', ]
paths_clean = ['dataset/generated/generated_pushkin.txt', 'dataset/generated/generated_curious_pushkin.txt']
for i in range(len(paths)):
    f =  open(paths[i], 'r', encoding='utf-8')
    text = f.read().lower()
    f.close()
    
    f = open(paths_clean[i], 'w', encoding='utf-8')
    cleaned_text = re.sub(r'[^а-яА-Я\s#.,!?;:\-—\'"«»()…]', '', text)
    print(cleaned_text, file=f)
    f.close()
```

```python
def draw_tokens_frequency(paths, vocab_size=1000, ylim=200, k=10, dataset_name='сгенерированном/оригинальном'):
    
    tokenizer = Tokenizer(BPE(unk_token='<unk>'))
    trainer = BpeTrainer(vocab_size=vocab_size, min_frequency=100, special_tokens=['<unk>', '<bos>', '<eos>', '<pad>', '<\n>'])
    tokenizer.pre_tokenizer = Whitespace()
    tokenizer.train(paths, trainer)

    tokens = []
    for path in paths:
        with open(path, 'r', encoding='utf-8') as f:
            tokens += tokenizer.encode(f.read()).tokens


    print(f'всего токенов: {len(tokens)}')
    print(f'среди них уникальных {len(set(tokens))}')
    
    counter = Counter(tokens)
    words_cou = sorted(i[1] for i in counter.items())
    
    ind_k = -1
    for i, v in enumerate(words_cou):
        if v >= k:
            ind_k = i
            break
            
    
    
    plt.figure(figsize=(16, 9))
    plt.ylim((0, ylim))
    plt.xlim((0, len(words_cou)))
    
    plt.plot(words_cou, linewidth=2, label='частотное распределение токенов')
    
    
    plt.axvline(x=ind_k, linestyle='--', color='purple', linewidth=2, alpha=0.3)
    plt.axhline(y=words_cou[ind_k], linestyle='--', color='purple', linewidth=2, alpha=0.3)
    
    
    plt.scatter(ind_k, words_cou[ind_k], color='r', s=100, zorder=5)
    
    plt.xlabel('количество токенов')
    plt.ylabel('частота токенов')
    
    
    
    plt.title(f'Размер словаря: {vocab_size}. В {dataset_name} датасете не менее {k} раз встречается {len(words_cou) - ind_k} ~ {(len(words_cou) - ind_k) / len(words_cou) * 100:.2f}% токенов')
    plt.legend()
    plt.show()
```

```python
paths = ['tokenizer/dataset/generated_pushkin.txt', 'tokenizer/dataset/generated_curious_pushkin.txt']
```

```python
for vocab_size in range(200, 2_000, 500):
    draw_tokens_frequency(paths=paths, vocab_size=vocab_size, ylim=100_000, k=1_000, dataset_name='сгенерированном')
```

## каким должен быть размер словаря?


1. достаточно большим, чтобы модель могла генерировать разнообразные тексты и не переобучалась
2. не слишком большим (основная масса токено встречалась 1000+ раз), чтобы модель выучилась


подберу оптимальный размер:

```python
for vocab_size in range(1300, 1700, 100):
    draw_tokens_frequency(paths=paths, vocab_size=vocab_size, ylim=10_000, k=1_000, dataset_name='сгенерированном')
```

довольно не плохие графики на размерах 1300-1500. Для того, чтобы лучше выбрать, нарисую график зависимости процента токенов 1000+ встречаний от размера словаря

```python
paths = ['tokenizer/dataset/generated_pushkin.txt', 'tokenizer/dataset/generated_curious_pushkin.txt']

vocab_sizes = []
perc = []
for s in tqdm(range(1300, 1500, 5)):
    vocab_sizes.append(s)

    tokenizer = Tokenizer(BPE(unk_token='<unk>'))
    trainer = BpeTrainer(vocab_size=s, min_frequency=100, special_tokens=['<unk>', '<bos>', '<eos>', '<pad>', '<\n>'])
    tokenizer.pre_tokenizer = Whitespace()
    tokenizer.train(paths, trainer)



    
    tokens = []
    for path in paths:
        with open(path, 'r', encoding='utf-8') as f:
            tokens += tokenizer.encode(f.read()).tokens

    counter = Counter(tokens)
    words_cou = sorted(i[1] for i in counter.items())
    
    ind_1000 = -1
    for i, v in enumerate(words_cou):
        if v >= 1000:
            ind_1000 = i
            break

    perc.append((len(words_cou) - ind_1000) / len(words_cou) * 100)
    
    
    
```

```python
plt.figure(figsize=(16, 9))
plt.plot(vocab_sizes, perc)
plt.title('процент токенов 1000+ встречаний в зависимотси от размера словаря')
plt.grid(True)
plt.xticks([i for i in range(1300, 1500, 10)])

plt.show()
```

видно, что окрестность 1410 - самый удачный размер с точки зрения выбора частоты токенов и увеличения размера словаря, тк с посел 1410 резко падает частота токенов в тексте, а до 1410 частота значительно не меняется, за то размер алфавита уменьшается, что не хорошо


мой выбор - словарь размером 1410 токенов

```python
def draw_tokens_frequency_per_tokenizer(paths, tokenizer,k=10, dataset_name='сгенерированном'):
    


    tokens = []
    for path in paths:
        with open(path, 'r', encoding='utf-8') as f:
            tokens += tokenizer.encode(f.read())


    print(f'всего токенов: {len(tokens)}')
    print(f'среди них уникальных {len(set(tokens))}')
    
    counter = Counter(tokens)
    words_cou = sorted(i[1] for i in counter.items())
    
    ind_k = -1
    for i, v in enumerate(words_cou):
        if v >= k:
            ind_k = i
            break
            
    
    
    plt.figure(figsize=(16, 9))
    plt.ylim((0, ylim))
    plt.xlim((0, len(words_cou)))
    
    plt.plot(words_cou, linewidth=2, label='частотное распределение токенов')
    
    
    plt.axvline(x=ind_k, linestyle='--', color='purple', linewidth=2, alpha=0.3)
    plt.axhline(y=words_cou[ind_k], linestyle='--', color='purple', linewidth=2, alpha=0.3)
    
    
    plt.scatter(ind_k, words_cou[ind_k], color='r', s=100, zorder=5)
    
    plt.xlabel('количество токенов')
    plt.ylabel('частота токенов')
    
    
    
    plt.title(f'Размер словаря: {vocab_size}. В {dataset_name} датасете не менее {k} раз встречается {len(words_cou) - ind_k} ~ {(len(words_cou) - ind_k) / len(words_cou) * 100:.2f}% токенов')
    plt.legend()
    plt.show()
```
